```python
import pandas as pd
data = pd.read_excel('Migrants_EducationLevel.xlsx')
```

```python
data.head()
data.info()
data.describe()
```

```python
data.dropna()
data.drop_duplicates()
```

```python
data.drop(columns=['region10_ori','region10_des','oecd_ori','oecd_des'], inplace=True)
```

```python
data.fillna('unknown', inplace=True)  # Replace NaN with 'unknown'
```

```python
data['CtryName_origin'] = data['CtryName_origin'].str.strip()
data['CtryName_destination'] = data['CtryName_destination'].str.strip()
data['isoo'] = data['isoo'].str.strip()
data['isod'] = data['isod'].str.strip()
data['sex'] = data['sex'].str.strip()
data['income'] = data['income'].str.strip()
data['income_des'] = data['income_des'].str.strip()
```

```python
data.rename(columns={'CtryName_origin': 'Country_of_Origin'}, inplace=True)
data.rename(columns={'CtryName_destination': 'Country_of_Destination'}, inplace=True)
data.rename(columns={'isoo': 'Country_of_Origin_ID'}, inplace=True)
data.rename(columns={'isod': 'Country_of_Destination_ID'}, inplace=True)
data.rename(columns={'year': 'Decade'}, inplace=True)
data.rename(columns={'sex': 'Gender'}, inplace=True)
data.rename(columns={'income': 'Country_of_Origin_Income'}, inplace=True)
data.rename(columns={'income_des': 'Country_of_Destination_Income'}, inplace=True)
data.rename(columns={'Tertiary_educated_migrants15': 'Formal_Education_Migrants'}, inplace=True)
data.rename(columns={'Non_tertiary_educated_migrants15': 'No_Formal_Education_Migrants'}, inplace=True)
data.rename(columns={'Children014': 'Migrant_Children'}, inplace=True)
data.rename(columns={'ForeignBorn': 'Total_Migrants'}, inplace=True)
```

```python
income_mapping = {
    'low income': 1,
    'lower middle income': 2,
    'upper middle income': 3,
    'high income': 4
}

data['Numerical_Country_of_Origin_Income'] = data['Country_of_Origin_Income'].map(income_mapping)
data['Numerical_Country_of_Destination_Income'] = data['Country_of_Destination_Income'].map(income_mapping)
```

```python
data.head()
data.info()
data.describe()
```

```python
data.to_csv('Migrants_EducationLevel CLEANED.csv', index=False)
```
